```python
from sklearn.model_selection import train_test_split
```

```python
x = df.drop("class", axis=1)
```

```python
x
```

```python
x.cap_shape=df.cap_shape.astype('category').cat.codes
x.cap_surface=df.cap_surface.astype('category').cat.codes
x.cap_color=df.cap_color.astype('category').cat.codes
x.gill_attachment=df.gill_attachment.astype('category').cat.codes
x.does_bruise_or_bleed=df.does_bruise_or_bleed.astype('category').cat.codes
x.gill_spacing=df.gill_spacing.astype('category').cat.codes
x.gill_color=df.gill_color.astype('category').cat.codes
x.stem_root=df.stem_root.astype('category').cat.codes
x.stem_surface=df.stem_surface.astype('category').cat.codes
x.stem_color=df.stem_color.astype('category').cat.codes
x.veil_type=df.veil_type.astype('category').cat.codes
x.veil_color=df.veil_color.astype('category').cat.codes
x.has_ring=df.has_ring.astype('category').cat.codes
x.ring_type=df.ring_type.astype('category').cat.codes
x.Spore_print_color=df.Spore_print_color.astype('category').cat.codes
x.habitat=df.habitat.astype('category').cat.codes
x.season=df.season.astype('category').cat.codes
x
```

```python
df_label = df["class"].copy()
```

```python
y=np.ravel(df_label)
```

```python
y
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.5, random_state=50)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler().fit(x_train)
x_train = scaler.transform(x_train)
x_test = scaler.transform(x_test)
```

```python
#https://scikit-learn.org/stable/modules/generated/sklearn.svm.SVC.html
```

```python
from sklearn.svm import SVC
svc_model = SVC()svc_model.fit(x_train, y_train)
```

```python
svc_model.fit(x_train, y_train)
```

```python
SVC(C=1.0, cache_size=200, class_weight=None, coef0=0.0,
decision_function_shape='ovr', degree=3, gamma='auto_deprecated',
kernel='rbf', max_iter=-1, probability=False, random_state=None,
shrinking=True, tol=0.001, verbose=False)
```

```python
y_predict = svc_model.predict(x_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
y_test
```

```python
y_predict
```

```python
cm = np.array(confusion_matrix(y_test, y_predict, labels=["e","p"]))
```

```python
confusion = pd.DataFrame(cm, index=['e', 'p'], columns=['Predicted edible', 'Predicted poisonous'])
```

```python
import seaborn as sns
sns.heatmap(confusion,annot=True,fmt='g')
```

```python
print(classification_report(y_test, y_predict))
```
